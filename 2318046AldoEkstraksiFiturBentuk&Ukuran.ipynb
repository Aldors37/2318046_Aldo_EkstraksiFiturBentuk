4. Melakukan ekstraksi fitur Merah, Hijau dan Biru pada 10 citra

```python
import cv2
import numpy as np
import os

# Koneksi ke Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Impor file gambar dari Google Drive
path = '/content/drive/MyDrive/Dataset Comvis'
gambar = []
for filename in os.listdir(path):
    img = cv2.imread(os.path.join(path, filename))
    if img is not None:
        gambar.append(img)

# Membuat variabel untuk menyimpan hasil ekstraksi fitur warna dan nama file
red_channel = []
green_channel = []
blue_channel = []
filenames = []

# Loop melalui setiap gambar dan menghitung nilai rata-rata untuk setiap saluran warna
for filename, img in zip(os.listdir(path), gambar):
    b, g, r = cv2.split(img)
    blue_channel.append(np.mean(b))
    green_channel.append(np.mean(g))
    red_channel.append(np.mean(r))
    filenames.append(filename)

# Menampilkan hasil ekstraksi fitur warna dan nama file
for i in range(len(gambar)):
    print("Gambar ke-", i+1, " (", filenames[i], ")")
    print("Rata-rata Saluran Warna Biru: ", blue_channel[i])
    print("Rata-rata Saluran Warna Hijau: ", green_channel[i])
    print("Rata-rata Saluran Warna Merah: ", red_channel[i])
    print()
```

5. Melakukan ekstraksi fitur luas pada 10 citra

```python
import cv2
import numpy as np
import os

# Koneksi ke Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Impor file gambar dari Google Drive
path = '/content/drive/MyDrive/Dataset Comvis'
gambar = []
for filename in os.listdir(path):
    img = cv2.imread(os.path.join(path, filename))
    if img is not None:
        gambar.append(img)

# Loop melalui setiap gambar dan menampilkan hasil ekstraksi fitur area
for i in range(len(gambar)):
    gray = cv2.cvtColor(gambar[i], cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cnt = max(contours, key=cv2.contourArea)
    area = cv2.contourArea(cnt)
    print("Gambar ke-", i+1, " (", os.listdir(path)[i], ")")
    print("Luas Area Objek: ", area)
    print()
```

6. Melakukan penampungan nilai-nilai fitur ke dalam file .xlsx atau .csv

```python
!pip install XlsxWriter

import cv2
import numpy as np
import os
import xlsxwriter

# Koneksi ke Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Impor file gambar dari Google Drive
path = '/content/drive/MyDrive/Dataset Comvis'
gambar = []
for filename in os.listdir(path):
    img = cv2.imread(os.path.join(path, filename))
    if img is not None:
        gambar.append(img)

# Buat file Excel dan header
nama_file = []
rata2_biru = []
rata2_hijau = []
rata2_merah = []
luas_area = []
workbook = xlsxwriter.Workbook('data_gambar.xlsx')
worksheet = workbook.add_worksheet()
header = ['Nama File', 'Rata-rata Warna Biru', 'Rata-rata Warna Hijau', 'Rata-rata Warna Merah', 'Luas Area Objek']
for col in range(len(header)):
    worksheet.write(0, col, header[col])

# Loop melalui setiap gambar dan menyimpan data ke dalam file Excel
for i in range(len(gambar)):
    b, g, r = cv2.split(gambar[i])
    gray = cv2.cvtColor(gambar[i], cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cnt = max(contours, key=cv2.contourArea)
    area = cv2.contourArea(cnt)
    nama_file.append(os.listdir(path)[i])
    rata2_biru.append(np.mean(b))
    rata2_hijau.append(np.mean(g))
    rata2_merah.append(np.mean(r))
    luas_area.append(area)

# Menyimpan data ke dalam file Excel
for row in range(len(nama_file)):
    worksheet.write(row+1, 0, nama_file[row])
    worksheet.write(row+1, 1, rata2_biru[row])
    worksheet.write(row+1, 2, rata2_hijau[row])
    worksheet.write(row+1, 3, rata2_merah[row])
    worksheet.write(row+1, 4, luas_area[row])

workbook.close()
print("Data telah berhasil disimpan ke dalam file Excel!")
```
